# Bag of Words Example

This notebook demonstrates Bag of Words (BoW) representation on a small text corpus.
- `CountVectorizer` from scikit-learn converts text to a matrix of token counts.
- Sentences are tokenized, fitted to a vocabulary, and then transformed into BoW vectors.

```python
# Import the CountVectorizer class for Bag of Words.
from sklearn.feature_extraction.text import CountVectorizer
```

## Prepare the text data

Define a small set of example sentences for tokenization and encoding.

```python
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

## Create and fit the vectorizer

`CountVectorizer` learns the vocabulary and builds a BoW matrix.

```python
bow = CountVectorizer()
bow_matrix = bow.fit_transform(documents)
```

```python
# Print the learned vocabulary for reference.
print("Vocabulary:", bow.get_feature_names_out())
```

```python
# Display the BoW matrix as a dense array.
print("BOW Matrix:\n", bow_matrix.toarray())
```

## Transform sentences and summarize

Each sentence is converted into a BoW vector showing word frequencies.
The resulting output is a sparse matrix converted to dense for easy inspection.

**Summary:**
- Unique words form the vocabulary.
- BoW creates a frequency vector for each document.
- This notebook uses dense output for easier inspection in a notebook.
- For larger corpora, sparse matrices are more memory efficient.

```python
for i, sentence in enumerate(documents):
    print(f"\nSentence {i+1}: {sentence}")
    print("BoW Vector:", bow_matrix[i].toarray()[0])
```
