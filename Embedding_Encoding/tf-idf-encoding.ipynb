# TF-IDF Encoding Example

This notebook demonstrates TF-IDF (Term Frequency-Inverse Document Frequency) encoding on a small text corpus using scikit-learn's `TfidfVectorizer`.

TF-IDF is a statistical measure that evaluates the importance of a word in a document relative to a collection of documents. It combines term frequency (TF) and inverse document frequency (IDF) to highlight unique terms while down-weighting common ones.

**Key Concepts:**
- **Term Frequency (TF):** Frequency of a term in a document.
- **Inverse Document Frequency (IDF):** Logarithm of total documents divided by documents containing the term.
- **TF-IDF Score:** TF * IDF, balancing local and global importance.

This example uses a small corpus to illustrate the process, with dense output for clarity.

```python
# Import the TfidfVectorizer from scikit-learn for TF-IDF computation
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Sample documents for demonstration
documents = [
    "people watch movie and watch movie again", 
    "people watch cricket and watch cricket",
    "people like movie and like movie a lot",
    "people like cricket"
]
```

## Prepare the Text Data

We define a small corpus of documents. Each document is a string representing a sentence about people's interests in movies and cricket. This allows us to see how TF-IDF weights terms based on frequency and rarity.

```python
# Initialize the TF-IDF vectorizer
tf_idf = TfidfVectorizer()
```

## Initialize the TF-IDF Vectorizer

Create an instance of `TfidfVectorizer`. By default, it handles tokenization, lowercasing, and stop word removal. We'll fit it to our documents to learn the vocabulary and IDF values.

```python
# Fit the vectorizer to the documents and transform them into TF-IDF vectors
tf_idf_vector = tf_idf.fit_transform(documents)
```

## Fit and Transform the Documents

Fit the vectorizer to the corpus to build the vocabulary and compute IDF. Then transform the documents into TF-IDF vectors. The result is a sparse matrix for efficiency.

```python
# Convert the sparse TF-IDF matrix to a dense array for easier viewing
tf_idf_vector.toarray()
```

## Convert to Dense Array for Inspection

The TF-IDF matrix is sparse by default. Convert to a dense NumPy array to view the full vectors easily in the notebook.

```python
# Print each TF-IDF vector for inspection
for vector in tf_idf_vector.toarray():
    print(vector)
```

## Display TF-IDF Vectors

Print each document's TF-IDF vector. Notice how terms like "movie" and "cricket" get weighted based on their frequency and rarity across documents.

```python
# Get the feature names (vocabulary) from the vectorizer
tf_idf.get_feature_names_out()
```

## Inspect the Vocabulary

Retrieve the feature names (unique terms) learned by the vectorizer. This shows the vocabulary in the order of the vectors.

```python
print("Vocabulary:", tf_idf.get_feature_names_out())
     
```
