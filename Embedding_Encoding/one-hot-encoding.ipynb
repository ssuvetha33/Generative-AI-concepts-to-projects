# One-Hot Encoding Example

This notebook demonstrates one-hot encoding on a small text corpus.
- `OneHotEncoder` from scikit-learn maps each unique word to a binary vector.
- Sentences are tokenized, fitted to an encoder vocabulary, and then transformed.

```python
# Import only the required class for one-hot encoding.
from sklearn.preprocessing import OneHotEncoder
```

## Prepare the text data

Define a small set of example sentences for tokenization and encoding.

```python
documents = [
    "people watch movie",
    "people watch cricket",
    "people like movie",
    "people like cricket"
]
```

## Tokenize the sentences

Convert each sentence to lowercase and split into individual words.

```python
tokens = [sentence.lower().split() for sentence in documents]
```

```python
# Display the tokenized sentences.
tokens
```

## Prepare input for the encoder

`OneHotEncoder` expects a 2D array where each row is one categorical value.

```python
all_words = [[word] for sentence in tokens for word in sentence]
```

```python
# Show the flattened list of single-word records.
all_words
```

## Fit the encoder

Learn the vocabulary of unique words from the training records.

```python
encoder = OneHotEncoder(sparse_output=False)
encoder.fit(all_words)
```

```python
# Print the learned vocabulary for reference.
print("Vocabulary:", encoder.categories_[0])
```

## Transform sentences and summarize

Each word in a sentence is converted into a one-hot vector.
The resulting output is a dense NumPy array with one row per word.

**Summary:**
- Unique words are extracted from the corpus.
- One-hot encoding creates a binary vector for each word.
- This notebook uses dense output for easier inspection in a notebook.
- For larger vocabularies, sparse matrices are more memory efficient.

```python
for sentence in tokens:
    encoded_sentence = encoder.transform([[word] for word in sentence])
    print("\nSentence:", sentence)
    print(encoded_sentence)
```
