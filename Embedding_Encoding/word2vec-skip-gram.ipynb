# Word2Vec Skip-Gram Example

This notebook demonstrates Skip-Gram using Gensim's `Word2Vec`.
- Skip-Gram predicts surrounding context words from the current word.
- The notebook also shows how to load Google's pre-trained Word2Vec vectors with `gensim.downloader`.

```python
from gensim.models import Word2Vec, KeyedVectors
import gensim.downloader as api
```

## Prepare the training corpus

Use a small tokenized corpus so the skip-gram behavior is easy to inspect.

```python
sentences = [
    ["people", "watch", "movie", "again"],
    ["people", "watch", "cricket", "again"],
    ["people", "like", "movie", "a", "lot"],
    ["people", "like", "cricket", "a", "lot"],
    ["movie", "fans", "like", "great", "stories"],
    ["cricket", "fans", "like", "great", "matches"]
]

sentences
```

## Train a Skip-Gram model

In Gensim, `sg=1` means Skip-Gram.

```python
skipgram_model = Word2Vec(
    sentences=sentences,
    vector_size=50,
    window=2,
    min_count=1,
    workers=1,
    sg=1,
    epochs=200
)
```

```python
print("Vocabulary:", list(skipgram_model.wv.index_to_key))
```

## Inspect word vectors

Skip-Gram usually does well when learning representations for less frequent words.

```python
print("Vector for 'cricket':\n", skipgram_model.wv["cricket"])
```

```python
print("Words similar to 'cricket':")
print(skipgram_model.wv.most_similar("cricket", topn=5))
```

## Load Google's pre-trained Word2Vec model

Google's published Word2Vec vectors are commonly used through Gensim as pre-trained `KeyedVectors`.
This download is large, so run it only when you want the pre-trained embeddings.

```python
google_news_vectors = api.load("word2vec-google-news-300")
isinstance(google_news_vectors, KeyedVectors)
```

```python
print("Vector size from Google News model:", google_news_vectors.vector_size)
print(google_news_vectors.most_similar(positive=["king", "woman"], negative=["man"], topn=5))
```

## Summary

- Skip-Gram predicts nearby words from the current word.
- It often learns stronger vectors for rare words than CBOW.
- Google's pre-trained Word2Vec vectors are useful when you want semantic relationships without local training.
